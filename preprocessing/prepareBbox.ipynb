```python
import pandas as pd
```

```python
df1 = pd.read_json("Final.json")
```

```python
df1
```

```python
for i in range(len(df1)) :
    bbox = df1.loc[i, "bbox"]
    h = df1.loc[i, "height"]
    w = df1.loc[i, "width"]
    
    width = bbox[2]
    height = bbox[3]
    
    x = bbox[0] + width / 2
    y = bbox[1] + height / 2
    
    
    size = height
    if width > height:
        size = width
    
    
    x1 = int(max(0, x - size))
    y1 = int(max(0, y - size))
    x2 = int(min(w, x + size))
    y2 = int(min(h, y + size))
    
    
    
    # print(bbox)
    # print(h, w)
    # print(list([x1, y1, x2, y2]))
    df1.at[i, "bbox"] = [x1, y1, x2, y2]

    # break
    # break
```

```python
df1
```

```python
df1 = df1.drop(['image_id', 'iscrowd', 'width', 'height', 'id'], axis=1)
```
