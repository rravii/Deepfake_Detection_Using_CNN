```python
from mtcnn import MTCNN
import os
import tensorflow as tf
import numpy as np
import cv2
from keras.models import load_model
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow
```

```python
base_path = "/content/drive/MyDrive/preprocessing"
```

```python
prediction_class = ["Fake", "Real"]
```

```python
detector = MTCNN()
```

#**Visualizing out test dataset**

```python
base_path_ds = "/content/drive/MyDrive/MinorProject/Some_images/test_dataset"
```

```python
def crop_img(img):

  bbox = detector.detect_faces(img)[0]['box']
  
  # print(bbox, img.shape)

  h = img.shape[0]
  w = img.shape[1]
  
  width = bbox[2]
  height = bbox[3]
  
  x = bbox[0] + width / 2
  y = bbox[1] + height / 2
  
  
  size = height * 0.8
  if width > height:
      size = width * 0.8
  
  
  x1 = int(max(0, x - size))
  y1 = int(max(0, y - size))
  x2 = int(min(w, x + size))
  y2 = int(min(h, y + size))
  

  newimg = img[y1:y2,x1:x2]
  
  # print([y1,y2,x1,x2])

  # cv2_imshow(img)
  # cv2_imshow(newimg)


  # print(newimg.shape)

  return newimg



  # # print(h, w)
  # # print(list([x1, y1, x2, y2]))
  # df1.at[i, "bbox"] = [x1, y1, x2, y2]
```

```python
def visualize(model_name):

  model_path = os.path.join(base_path, model_name)

  model = load_model(model_path)

  files = os.listdir(base_path_ds)

  print(len(files))

  plt.figure(figsize=(20, 20))

  i = 1
  for file in files:
    fig = plt.subplot(5,5,i)
    path = os.path.join(base_path_ds, file)
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # img = crop_img(img)
    img = cv2.resize(img, (256, 256))
    img = img / 255
    plt.imshow(img)
    pred = model.predict(np.expand_dims(img, axis=0))
    index = np.argmax(pred, axis = 1)
    name = file.split("_")[0]
    fig.text(0, -10, f"Actual: {name} \n Prediction: {prediction_class[index[0]]} {pred[0][index]}")
    plt.axis('off')
    i = i + 1
  plt.show()
```

```python

def visualize_real(model_name):

  model_path = os.path.join(base_path, model_name)

  model = load_model(model_path)
  
  files = os.listdir(base_path_world)

  print(len(files))
  plt.figure(figsize=(20, 20))



  i = 1
  for file in files:
    fig = plt.subplot(5,5,i)
    path = os.path.join(base_path_world, file)
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = crop_img(img)
    img = cv2.resize(img, (256, 256))
    img = img / 255
    plt.imshow(img)
    pred = model.predict(np.expand_dims(img, axis=0))
    index = np.argmax(pred, axis = 1)
    name = file.split("_")[0]
    fig.text(0, -10, f"Actual: {name} \n Prediction: {prediction_class[index[0]]} {pred[0][index]}")
    plt.axis('off')
    i = i + 1
  plt.show()
```
